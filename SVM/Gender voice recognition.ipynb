The following acoustic properties of each voice are measured and included within the CSV:
1. meanfreq: mean frequency (in kHz)
2. sd: standard deviation of frequency
3. median: median frequency (in kHz)
4. Q25: first quantile (in kHz)
5. Q75: third quantile (in kHz)
6. IQR: interquantile range (in kHz)
7. skew: skewness (see note in specprop description)
8. kurt: kurtosis (see note in specprop description)
9. sp.ent: spectral entropy
10. sfm: spectral flatness
11. mode: mode frequency
12. centroid: frequency centroid (see specprop)
13. peakf: peak frequency (frequency with highest energy)
14. meanfun: average of fundamental frequency measured across acoustic signal
15. minfun: minimum fundamental frequency measured across acoustic signal
16. maxfun: maximum fundamental frequency measured across acoustic signal
17. meandom: average of dominant frequency measured across acoustic signal
18. mindom: minimum of dominant frequency measured across acoustic signal
19. maxdom: maximum of dominant frequency measured across acoustic signal
20. dfrange: range of dominant frequency measured across acoustic signal
21. modindx: modulation index. Calculated as the accumulated absolute difference between adjacent measurements of fundamental frequencies divided by the frequency range
22. label: male or female

# No null value. So, we can go for SVM Algorithm

Separating Input features and output feature

```python
# Take out all feature except our Output feature =  'Label'
input_features=voice_df.iloc[:, :-1]
input_features.head()
```

Converting string value to int type for labels

```python
# Take our our output feature == 'Label'
output=voice_df.iloc[:,-1]
output.sample(5)
```

```python
# label has 'male' and 'female'. So, encoding label
output_encoder = LabelEncoder()
y = output_encoder.fit_transform(output)
# From above cell at 2021 and 704 we have female and male
print("female is encode by {} and male encoded by {}".format(y[2021],y[704]))
print("Below random sample of y")
print(random.choice(y,5))
```

Here our input feature highly vary in magnitudes, units, and range, So better to scale all these

```python
sc_X = StandardScaler()
X= pd.DataFrame(sc_X.fit_transform(input_features),columns=input_features.columns,
                          index = input_features.index)
```

Splitting dataset into training set and testing set for better generalisation

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)
```

Running SVM with default hyperparameter, default kernel = 'rbf'.
    "kernel methods are a class of algorithms for pattern analysis"

```python
svc=SVC()
svc.fit(X_train,y_train)
```

Kernel = rbf accuracy

```python
y_pred=svc.predict(X_test)
rbf_acc = metrics.accuracy_score(y_test,y_pred);
print("RBF Accuracy {}".format(rbf_acc))
```

Kernel = Linear

```python
svc=SVC(kernel='linear')
svc.fit(X_train,y_train)
y_pred=svc.predict(X_test)
linear_acc = metrics.accuracy_score(y_test,y_pred);
print("Linear Accuracy {}".format(linear_acc))
```

Kernel = Polynomial

```python
svc=SVC(kernel='poly')
svc.fit(X_train,y_train)
y_pred=svc.predict(X_test)
polynomial_acc = metrics.accuracy_score(y_test,y_pred);
print("Polynomial Accuracy {}".format(polynomial_acc))
```

Moving from train_test_split to K-fold(Cross Validation = CV) because TTS are some what bias. Performing K-fold cross validation with different kernels. 
1. https://stackoverflow.com/a/49134560

kernel = rbf

```python
svc=SVC(kernel='rbf')
rbf_acc_cv = cross_val_score(svc, X, y, cv=10, scoring='accuracy')
print(rbf_acc_cv)
print("Average of rbf with cv {}  and rbf accuracy without cv {}".format(rbf_acc_cv.mean(), rbf_acc))
```

CV on Linear kernel

```python
svc=SVC(kernel='linear')
linear_acc_cv = cross_val_score(svc, X, y, cv=10, scoring='accuracy')
print(linear_acc_cv)
print("Average of linear with cv {}  and linear accuracy without cv {}".format(linear_acc_cv.mean(), linear_acc))
```

CV on Polynomial kernel

```python
svc=SVC(kernel='poly')
polynomial_acc_cv = cross_val_score(svc, X, y, cv=10, scoring='accuracy')
print(polynomial_acc_cv )
print("Average of Polynomial with cv {}  and Polynomial accuracy without cv {}".format(polynomial_acc_cv .mean(), polynomial_acc))
```

So, By default for all kernel cross validation giving less accuracy score

Now Tweaking parameters

Taking all the values of C and checking out the accuracy score with kernel as linear.
https://stats.stackexchange.com/a/31067

```python
C_range=list(range(1,51))
acc_score_c_change=[]
for c in C_range:
    svc = SVC(kernel='linear', C=c)
    score_c = cross_val_score(svc, X, y, cv=10, scoring='accuracy')
    acc_score_c_change.append((score_c.mean(),c))
print("Accuracy Score List:")
print(acc_score_c_change)
```

```python
print("Max Score with c : {} , Min Score with c: {}".format(max(acc_score_c_change,key=lambda item:item[0]),min(acc_score_c_change,key=lambda item:item[0]) )) 
```

```python
gamma_range=[0.0001,0.001,0.01,0.1,1,1.5,2,2.5,3]
acc_score_gaama_change=[]
for g in gamma_range:
    svc = SVC(kernel='linear', gamma=g)
    score_gaama = cross_val_score(svc, X, y, cv=10, scoring='accuracy')
    acc_score_gaama_change.append((score_gaama.mean(),g))
print("Accuracy Score List:")
print(acc_score_gaama_change)
```

```python
print("Max Score with gaama : {} , Min Score with gaama: {}".format(max(acc_score_gaama_change,key=lambda item:item[0]),min(acc_score_gaama_change,key=lambda item:item[0])))
```

```python
degree= list(range(1,10,1))
acc_score_degree_change=[]
for d in degree:
    svc = SVC(kernel='linear', degree=d)
    score_degree = cross_val_score(svc, X, y, cv=10, scoring='accuracy')
    acc_score_degree_change.append((score_degree.mean(),d))
print("Accuracy Score List:")
print(acc_score_degree_change)  
 
```

```python
print("Max Score with degree : {} , Min Score with degree: {}".format(max(acc_score_degree_change,key=lambda item:item[0]),min(acc_score_degree_change,key=lambda item:item[0]))) 
```

Now using best c = 1 , gamma = 0.0001 and degree = 1 in linear kernel

```python
svc = SVC(C=1,kernel='linear', degree=d , gamma = 0.0001 )
best_score = cross_val_score(svc, X, y, cv=10, scoring='accuracy')
print("score with best parameter  {}  and score with default parameter {}".format(best_score.mean(),linear_acc_cv.mean()))
```
